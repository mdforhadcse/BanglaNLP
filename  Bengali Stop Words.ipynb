```python
import pandas as pd
```

```python
def BnStopWordRemover(text):
  import requests 
  stopwordsraw = requests.get('https://raw.githubusercontent.com/FazlyRabbiBD/NLP/main/BnStopwordsTest2.txt') 
  steiped=stopwordsraw.text.strip("")
  separated=steiped.splitlines()
  BnStopWordList=list(separated)
  for i in BnStopWordList:
    n=text.replace(i,'')
    text=n
  return text
p1='রবীন্দ্রনাথ অন্তত এখানে কবে খেতে অথচ অনেক অবশ্য এসেছিল অথচ?'
BnStopWordRemover(p1)
data = {'Headline':['রবীন্দ্রনাথ অন্তত এখানে কবে খেতে অথচ', 
                    'তীব্র স্রোতে বাংলাবাজার-শিমুলিয়া নৌপথে বন্ধ হল', 
                    'বাংলাবাজার-শিমুলিয়ায় চলছে ৫ ফেরি, পারাপারে লাগ', 
                    'পদ্মা সেতুর ১০ নম্বর পিলারে আবারও ফেরির ধাক্কা']}
 
# Create DataFrame
df = pd.DataFrame(data)
df['X'] = df.apply(lambda row : BnStopWordRemover(row['Headline']), axis = 1)
```
